```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from textblob import TextBlob
```

```python
vader = SentimentIntensityAnalyzer()

def vader_compound(text):
    return vader.polarity_scores(text)["compound"]

def textblob_polarity(text):
    return TextBlob(text).sentiment.polarity

df["vader"] = df["review"].apply(vader_compound)
df["textblob"] = df["review"].apply(textblob_polarity)
```

```python
df[["vader","textblob","label"]].head()
```

```python
import numpy as np

X = df[["vader","textblob"]].values.astype(np.float32)
y = df["label"].values.astype(np.float32).reshape(-1,1)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
np.random.seed(42)

indices = np.arange(len(X))
np.random.shuffle(indices)

split = int(0.8 * len(X))

train_idx = indices[:split]
test_idx = indices[split:]

X_train = X[train_idx]
y_train = y[train_idx]

X_test = X[test_idx]
y_test = y[test_idx]

print(X_train.shape, X_test.shape)
```

```python
import sys
sys.path.append("src")
```

```python
from mlp import MLP
```

```python
model = MLP(input_size=2, hidden_size=8, lr=0.5)
```

```python
model.train(X_train, y_train, epochs=50)
```

```python
pred = model.predict(X_test)

accuracy = (pred == y_test).mean()
print("Test Accuracy:", accuracy)
```

```python
import numpy as np

np.savez(
    "model/mlp_weights.npz",
    W1=model.W1,
    b1=model.b1,
    W2=model.W2,
    b2=model.b2
)

print("Model saved!")
```
